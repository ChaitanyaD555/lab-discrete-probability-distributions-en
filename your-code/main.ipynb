## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
# What is the probability that the first 5 fruits are all apples and the 15 fruits are all oranges in a random set ?
n1 = 5  # Number of apples
n2 = 15 # Number of oranges
p = 0.6  # Probability of apple
q = 0.4  # Probability of orange   

probability_all_apples_and_oranges = (p ** n1) * (q ** n2)
print(f"Probability that the first {n1} fruits are all apples and the next {n2} fruits are all oranges: {probability_all_apples_and_oranges:.8f}")
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
#In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?
cdf_less_than_5_apples = stats.binom.cdf(5, n1 + n2, p)
print(f"Probability that less than 5 fruits picked are apples: {cdf_less_than_5_apples:.4f}")   
# In the lesson we also discussed PMF (probability mass function). In the context of the above problem, **what is the probability that exactly 5 fruits picked are apples?
pmf_exactly_5_apples = stats.binom.pmf(5, n1 + n2, p)
print(f"Probability that exactly 5 fruits picked are apples: {pmf_exactly_5_apples:.4f}")
````

<!-- **Plot the PDF (probability distributiom function) for the above binomial distribution.** -->

**Plot the Probability Mass Function (PMF) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
# your code here
# Please label the axes and give a title to the plot 
plt.figure(figsize=(10, 6))
x = np.arange(0, n1 + n2 + 1)
y = stats.binom.pmf(x, n1 + n2, p)
plt.stem(x, y, basefmt=" ")
plt.xlabel('Number of Apples in Sample of 20')
plt.ylabel('Probability')
plt.title('Binomial Distribution PMF (n=20, p=0.6)')
plt.xticks(x)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of
```

````python
import math
math.factorial(n)

# poissons parametes
mu = 2.3 # average number of goals
k=5

# Poisson probability formula: P(X=k) = (e^-mu) * (mu^k) / k!   
prob_5_goals = (math.exp(-mu) * (mu ** k)) / math.factorial(k)
print(f"Probability that the number of goals scored in a match is 5: {prob_5_goals:.4f}")
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
# your code here
# Please label the axes and give a title to the plot 

mu = 2.3  # average number of goals

x = np.arange(0, 11)  # 0 to 10 goals
y = [((math.exp(-mu) * (mu ** k)) / math.factorial(k)) for k in x]

plt.figure(figsize=(8, 5))
plt.stem(x, y, basefmt=" ")
plt.xlabel('Number of Goals')
plt.ylabel('Probability')
plt.title('Poisson Distribution PMF (mu=2.3) for Goals 0-10')
plt.xticks(x)
plt.grid(axis='y', linestyle='--', linewidth=1.0)
plt.show()
````
